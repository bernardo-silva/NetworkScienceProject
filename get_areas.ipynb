```python
import requests as r
from bs4 import BeautifulSoup
from time import sleep
```

```python
area_codes = [14,21,22,31,32,34,38,42,44,46,48,52,54,
             58,62,64,72,76,81,84,85,86,99]
url = "https://www.dges.gov.pt/guias/indarea.asp?area="
```

```python
areas = []
area_dict  = {}
for code in area_codes:
    result = r.get(url+str(code))
    soup = BeautifulSoup(result.content,"html5lib")
    p = soup.findAll("p",attrs={"class":"box10"})
    courses = soup.findAll("div",attrs={"class":"lin-area-c2"})
    area = str(p).split(">")[1].split("<")[0]
    
    area_dict[area] = [str(x.findChildren("a")).split(">")[1].split("<")[0] for x in courses]
    # print([str(x).split(">")[1].split("<")[0] for x in courses])
    print(courses[0].findChildren("a"))
    # print(str(p).split(">")[1].split("<")[0])
    areas.append(area)
    sleep(0.2)
```

```python
area_dict["Humanidades"] += area_dict[areas[-1]]
area_dict.pop(areas[22])

area_dict["Ciências da Vida"] += area_dict[areas[20]]
area_dict.pop(areas[20])

area_dict["Ciências Empresariais"] += area_dict[areas[19]]
area_dict.pop(areas[19])

area_dict["Engenharia e Técnicas Afins"] += area_dict[areas[10]]
area_dict["Engenharia e Técnicas Afins"] += area_dict[areas[8]]
area_dict.pop(areas[8])
area_dict.pop(areas[10])
print(area_dict.keys)
```
