```python
import numpy as np
import pandas as pd
import json
```

```python
year = 2018
data = pd.read_excel(f'data/colocados_{year}.xls',header=4)
data = data[["Código Instit.","Código Curso","Vagas Iniciais","Colocados ","Nota do últ. colocado (cont. geral)"]]
data = data.drop([0,1069,1070])
# print(data.keys())
data
```

```python
#Load data set
with open(f"courses_{year}.json","r") as f:
    courses = json.load(f)
```

```python
for c in courses:
    a = data[data["Código Curso"] == c["code"]]
    # print(a)
    a = a[a["Código Instit."] == c["faculty"]['code']]
    if not len(a): print("Ups")
    # print(a['Nota do últ. colocado (cont. geral)'])
    c['Accepted'] = int(a['Colocados '])
    c['vagas'] = int(a['Vagas Iniciais'])
    c['grade'] = float(a['Nota do últ. colocado (cont. geral)'])
    c['grade'] = c['grade'] if not np.isnan(c['grade']) else 0
```

```python
#Save data as json file
with open(f"courses_{year}.json","w", encoding='utf8') as f:
    json.dump(courses,f, ensure_ascii=False)
```
